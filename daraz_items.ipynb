```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://ultima.com.np/"

headers={
    "User-Agent": "My Web Scraper 1.0 - for educational purposes"
}

response=requests.get(url,headers=headers)
```

```python
soup=BeautifulSoup(response.text,'lxml')
print(soup)
```

```python
#we can use .prettify() to see the HTML in a nicely indented format
print(soup.prettify())

#--Finding Elements--
#Find the first element of a certain tag (eg., the page title)

page_title=soup.find('title')
print(f"Page Title: {page_title.text}")
```

```python
titles=soup.find_all('div',class_='relative')
print(titles)
```

```python
daraz_titles=[]
for title in titles:
    daraz_items=title.find('div')
    daraz_items_title=daraz_items.text
    item_details={
        'Title':daraz_items_title
    }
    daraz_titles.append(item_details)
print(daraz_titles)
```
