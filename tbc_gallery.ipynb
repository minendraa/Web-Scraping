```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.thebritishcollege.edu.np/gallery"

headers={
    "User-Agent": "My Web Scraper 1.0 - for educational purposes"
}

response=requests.get(url,headers=headers)
```

```python
soup=BeautifulSoup(response.text,'lxml')
print(soup)
```

```python
#we can use .prettify() to see the HTML in a nicely indented format
print(soup.prettify())

#--Finding Elements--
#Find the first element of a certain tag (eg., the page title)

page_title=soup.find('title')
print(f"Page Title: {page_title.text}")
```

```python
titles=soup.find_all('div',class_='n-wrap')
print(titles)
```

```python
tbc_gallery=[]
for title in titles:
    tbc_photo_title=title.find('h4')
    photo_title=tbc_photo_title.text
    photo_details={
        'Title':photo_title
    }
    tbc_gallery.append(photo_details)
print(tbc_gallery)
```
